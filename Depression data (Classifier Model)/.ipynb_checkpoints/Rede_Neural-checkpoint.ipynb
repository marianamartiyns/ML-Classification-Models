````python
X = df_Aptos.drop('Diagnóstico depressão', axis=1)
y = df_Aptos['Diagnóstico depressão']

# Substituindo o valor 2 por 0 na coluna 'Diagnóstico depressão' para evitar problemas nos modelos
y = y.replace(2, 0)
y
````

#### Normalização e/ou Padronização dos dados

É necessario normalizar os dados uma vez que os algoritmos como SVM e redes neurais são sensíveis à escala dos dados.
**Padronização (Z-score)** é mais apropriada se seus dados seguirem ou precisarem se aproximar de uma distribuição normal.


````python
colunas_normalizar =  ['Dias trabalho', 'Aval. Saúde', 'Bem estar', 'Frequência Problemas Sono', 'Frequência Indisposição',
                      'Frequência falta de prazer', 'Frequência falta concentração', 'Frequência problemas alimentação',
                      'Frequência Lentidão ou Agitação', 'Frequência sem perspectiva', 'Frequência sentimento de fracasso',
                      'Frequência ideação suicida', 'Pratica exercícios físicos?', 'Horas de tela', 'Fuma Tabaco atualmente?',
                      'Fumou no passado?', 'IMC']

scaler = StandardScaler()

# Ajustar e transformar o scaler apenas no conjunto de treino para as colunas selecionadas
X_train_scaled = scaler.fit_transform(X_train[colunas_normalizar])

# Convertendo de volta para DataFrame para preservar os nomes das colunas
X_train_scaled = pd.DataFrame(X_train_scaled, columns=colunas_normalizar, index=X_train.index)

# Substituindo as colunas normalizadas no DataFrame original de treino
X_train[colunas_normalizar] = X_train_scaled

# Para o conjunto de teste, apenas transformar, sem ajustar o scaler
X_test_scaled = scaler.transform(X_test[colunas_normalizar])

# Convertendo de volta para DataFrame e substituindo no conjunto de teste
X_test_scaled = pd.DataFrame(X_test_scaled, columns=colunas_normalizar, index=X_test.index)
X_test[colunas_normalizar] = X_test_scaled
````

## Rede Neural





Sabendo que uma ou duas camadas escondidas são suficientes, podemos calcular o números de neuronios para cada camada, utilizando a fórmula apresentada em sala:

### Cálculo do Número de Neurônios por Camada Escondida

Para calcular o número de neurônios por camada escondida, podemos usar a fórmula apresentada no slide:

n <= DS - 10/10(d + 2)

onde:
- \( DS \) é o número de exemplos no dataset.
- \( d \) é o número de características (features) na camada de entrada.
- \( n \) é o número de neurônios na camada escondida.

````python
# calculo o número de neuronios para cada camada
def qnt_neuro(qnt_ex,d):
  n = (qnt_ex - 10) / (10*(d+2))
  return int(n)

qnt_ex = X_rn.shape[0]
d = X_rn.shape[1]

print(f'Quantidade de exemplos: {qnt_ex}')
print(f'Quantidade de features: {d}')

qnt_neuronios = qnt_neuro(qnt_ex,d)
print(f'\nNeurônios por camada: {qnt_neuronios}')
````

Pela teoria, a dimensão VC é aproximadamente igual ao número de parâmetros.

Sabendo disso calculamos:

````python
def calcular_dVC(qnt_neuro, d, saida=1):

    # Pesos da camada escondida
    pesos_escondidos = (d + 1) * qnt_neuro

    # Pesos da camada de saída
    pesos_saida = (qnt_neuro + 1) * saida

    # Dimensão VC aproximada (número total de pesos)
    dVC = pesos_escondidos + pesos_saida
    return dVC

d = 23  # Número de features
qnt_neuronios = 51  # Número de neurônios na camada escondida
dimensao_VC = calcular_dVC(qnt_neuronios, d)
print(f"Dimensão VC: {dimensao_VC}")
````

Verificação da Regra de Ouro: Sugere que a dimensão VC deve ser menor do que o número de exemplos para garantir uma boa capacidade de generalização.

Com isso, calculamos a quantidade recomendada de exemplos de treino com base na Regra de Ouro.

Essa abordagem considera que, para uma boa generalização, o número de exemplos de treino deve ser pelo menos 10 vezes a dimensão VC.

````python
def regra_ouro(dVC):
    return 10 * dVC

qnt_ex = 12892  # Quantidade de exemplos disponível no treino
print("Quantidade de exemplos para treino que temos: ", qnt_ex)
print("Quantidade de exemplos para treino que precisamos: ", regra_ouro(dimensao_VC))
````

````python
# Criando a arquitetura da rede neural com weight decay
modelo_decay = Sequential()
modelo_decay.add(Input(shape=(X_train.shape[1],)))
modelo_decay.add(Dense(285, kernel_initializer='normal', activation='relu', kernel_regularizer=regularizers.l2(0.01) ))  # Camada oculta 1
modelo_decay.add(Dense(1, kernel_initializer='normal', activation='sigmoid'))  # Camada de saída

# Configuração do otimizador
sgd = optimizers.SGD(learning_rate=0.01)

# Compila o modelo
modelo_decay.compile(loss='binary_crossentropy', optimizer=sgd, metrics=['accuracy'])

print(modelo_decay.summary())
````

````python
# Checar novamente as classes únicas após a conversão
classes_unicas = np.unique(y_train)
print("Classes presentes em y_train:", classes_unicas)

# Calcular pesos das classes
class_weights = compute_class_weight(class_weight='balanced', classes=classes_unicas, y=y_train)
class_weights_dict = {int(classe): peso for classe, peso in zip(classes_unicas, class_weights)}

# Verificar o tipo de dados e conteúdo
print("Tipo de classes únicas:", classes_unicas.dtype)
print("Tipo de chaves no dicionário de pesos:", type(list(class_weights_dict.keys())[0]))
print("Pesos calculados para as classes:", class_weights_dict)

# Treinar o modelo com os pesos de classes calculados
history_decay = modelo_decay.fit(X_train, y_train, epochs=100, batch_size=16, class_weight=class_weights_dict)
````

````python
# Plotando a métrica de erro ao longo das épocas
plt.plot(history_decay.history['loss'])
plt.title('Métrica de Erro ao Longo das Épocas')
plt.ylabel('Erro')
plt.xlabel('Época')
plt.legend(['Treinamento'])
plt.show()

# Avaliando a acurácia usando accuracy_score() nos dados de treino
pred = np.round(modelo_decay.predict(X_train))
acc_train = accuracy_score(y_train, pred)

# Avaliando nos dados de teste
E_out, acc_test = modelo_decay.evaluate(X_test, y_test, batch_size=60)

print(f'\n--> Acurácia (treino): {acc_train:.4f}')
print(f'--> Acurácia (teste): {acc_test:.4f}')
print(f"--> E_out - E_in = {E_out - history_decay.history['loss'][-1]:.4f}")
print(f"--> acc_train - acc_test = {acc_train - acc_test:.4f}")
````

````python
# Previsões e Avaliação do Modelo
y_train_pred = modelo_decay.predict(X_train) > 0.5
y_test_pred = modelo_decay.predict(X_test) > 0.5

# Calculando métricas
Ein = 1 - accuracy_score(y_train, y_train_pred)
Eout = 1 - accuracy_score(y_test, y_test_pred)

print(f'Ein: {Ein:.4f}')
print(f'Eout: {Eout:.4f}')

# Relatório de classificação
print("\nRelatório de classificação (dados de teste):")
print(classification_report(y_test, y_test_pred.astype(int)))
````
