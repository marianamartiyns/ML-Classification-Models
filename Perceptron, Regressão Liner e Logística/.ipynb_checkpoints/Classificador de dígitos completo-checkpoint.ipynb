# Classificador de d√≠gitos completo

#### Estrat√©gia "Um Contra Todos"
Classifica√ß√£o Bin√°ria para cada d√≠gito: Para cada d√≠gito ùëñ, cria-se um classificador bin√°rio que tenta distinguir o d√≠gito ùëñ dos demais.

Para o d√≠gito 0: ùë¶ = +1 para 0, y=‚àí1 para 1, 4 e 5. <br>
Para o d√≠gito 1: ùë¶ = +1 para 1, y=‚àí1 para 4 e 5 (0 √© removido). <br>
Para o d√≠gito 4: ùë¶ = +1 para 4, y=‚àí1 para 5 (0 e 1 s√£o removidos). <br>

```python
def plot_simetria_intensidade(caminho_treino, caminho_teste):
    # Carrega os datasets reduzidos
    df_treino = pd.read_csv(caminho_treino, delimiter=';')
    df_teste = pd.read_csv(caminho_teste, delimiter=';')
    
    # Cria arrays para simetria e intensidade
    X_treino = df_treino[['simetria', 'intensidade']].values
    y_treino = df_treino['label']
    
    X_teste = df_teste[['simetria', 'intensidade']].values
    y_teste = df_teste['label']

    # Plotar gr√°ficos
    plt.figure(figsize=(14, 6))

    # Gr√°fico de treino
    plt.subplot(1, 2, 1)
    scatter_treino = plt.scatter(X_treino[:, 0], X_treino[:, 1], c=y_treino, cmap=plt.cm.RdYlBu, edgecolor='k', s=40, alpha=0.8)
    plt.title('Dados de Treino')
    plt.xlabel('Simetria')
    plt.ylabel('Intensidade')
    plt.colorbar(scatter_treino, ticks=np.unique(y_treino), label='Classe')

    # Gr√°fico de teste
    plt.subplot(1, 2, 2)
    scatter_teste = plt.scatter(X_teste[:, 0], X_teste[:, 1], c=y_teste, cmap=plt.cm.RdYlBu, edgecolor='k', s=40, alpha=0.8)
    plt.title('Dados de Teste')
    plt.xlabel('Simetria')
    plt.ylabel('Intensidade')
    plt.colorbar(scatter_teste, ticks=np.unique(y_teste), label='Classe')

    plt.tight_layout()
    plt.show()
    
    return df_treino, df_teste, X_treino, X_teste, y_treino, y_teste
    
    
# Exemplo de como usar a fun√ß√£o
df_treino, df_teste, X_treino, X_teste, y_treino, y_teste = plot_simetria_intensidade('dataset_digits/train_redu.csv', 'dataset_digits/test_redu.csv')
```

```python
# Fun√ß√£o para classificar um novo exemplo
def classificar_digito(x, modelos, scaler_modelos='', model_type='linear'):
    x = np.array(x).reshape(1, -1)  # Reformatar para um array 2D

    # Inicializar vari√°veis para armazenar as predi√ß√µes
    pred_f0, pred_f1, pred_f4 = None, None, None

    if model_type == 'linear' or model_type == 'logistic':
        # Para Regress√£o Linear e Log√≠stica, utilizamos a fun√ß√£o `predict`
        pred_f0 = modelos[0].predict(x)
        pred_f1 = modelos[1].predict(x)
        pred_f4 = modelos[2].predict(x)
    
    elif model_type == 'perceptron':
        # Para Perceptron, utilizamos uma fun√ß√£o similar ao produto escalar
        
        # Normalizar o novo exemplo com os scalers correspondentes
        x_normalized_f0 = scaler_modelos[0].transform(x)
        x_normalized_f1 = scaler_modelos[1].transform(x)
        x_normalized_f4 = scaler_modelos[2].transform(x)

        # Fazer previs√µes usando os modelos
        pred_f0 = classify(x_normalized_f0, modelos[0])
        pred_f1 = classify(x_normalized_f1, modelos[1])
        pred_f4 = classify(x_normalized_f4, modelos[2])

    # Decis√£o com base nas predi√ß√µes
    if pred_f0 > 0:
        return 0
    elif pred_f1 > 0:
        return 1
    elif pred_f4 > 0:
        return 4
    else:
        return 5
```

### 3. Separa√ß√£o de Recursos e Labels
Separe os recursos (intensidade, simetria) e as labels para os conjuntos de treinamento e teste.

```python
def criar_dataset_binario(df, digito, excluir_labels=None):
    """
    Transforma um dataset multiclasse em um dataset bin√°rio para um d√≠gito espec√≠fico,
    com a op√ß√£o de excluir certos labels.
    
    Args:
    - df (DataFrame): O dataset original com labels multiclasse.
    - digito (int): O d√≠gito para o qual se deseja criar o dataset bin√°rio.
    - excluir_labels (list): Lista de labels a serem exclu√≠dos do dataset.
    
    Returns:
    - DataFrame: O dataset bin√°rio com o d√≠gito como +1 e todos os outros como -1, 
                 excluindo os labels especificados.
    """
    if excluir_labels is None:
        excluir_labels = []

    df_binario = df.copy()
    df_binario = df_binario[~df_binario['label'].isin(excluir_labels)]  # Excluir labels
    df_binario['label'] = df_binario['label'].apply(lambda x: 1 if x == digito else -1)
    return df_binario
```

```python
# Criar datasets bin√°rios para treino
df_0_x_todos_treino = criar_dataset_binario(df_treino, 0)

# Criar dataset bin√°rio para o d√≠gito 1, excluindo label 0
df_1_x_todos_treino = criar_dataset_binario(df_treino, 1, excluir_labels=[0])

# Criar dataset bin√°rio para o d√≠gito 4, excluindo labels 0 e 1
df_4_x_todos_treino = criar_dataset_binario(df_treino, 4, excluir_labels=[0,1])


# Exibindo os datasets bin√°rios criados (opcional)
print("Dataset 0 x Todos - Treino")
print(df_0_x_todos_treino[df_0_x_todos_treino['label'] == 1])

# Dataset 1 x Todos - Treino
print("\nDataset 1 x Todos - Treino")
print(df_1_x_todos_treino[df_1_x_todos_treino['label'] == 1])

# Dataset 4 x Todos - Treino
print("\nDataset 4 x Todos - Treino")
print(df_4_x_todos_treino[df_4_x_todos_treino['label'] == 1])
```

### 4.1 Perceptron

Essa √© a distibui√ß√£o dos nossos dados inicialmente:

```python
print("\nDistribution of Labels in Train Data:")
print(df_train_redu['label'].value_counts())

print("\nDistribution of Labels in Test Data:")
print(df_test_redu['label'].value_counts())
```

A distribui√ß√£o desses dados de treinamento e teste mostra que h√° um desequil√≠brio nas quantidades de exemplos para cada classe, especialmente com o d√≠gito 0, que tem significativamente menos exemplos do que os outros d√≠gitos. Esse desequil√≠brio pode estar afetando o desempenho do perceptron. Ai uma das estrategias √© a Valida√ß√£o Cruzada: garante que a divis√£o dos dados de treino e teste mantenha a propor√ß√£o original das classes. Isso ajuda a garantir que todas as classes estejam representadas de maneira equilibrada em cada subconjunto. 

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.preprocessing import StandardScaler

# Fun√ß√£o para executar a valida√ß√£o cruzada estratificada
def valida_cruzada_estratificada(X, y, max_it, n_splits=5):
    skf = StratifiedKFold(n_splits=n_splits)
    accuracies = []
    
    for train_index, test_index in skf.split(X, y):
        X_train, X_test = X[train_index], X[test_index]
        y_train, y_test = y[train_index], y[test_index]
        
        # Normalizar os dados
        scaler = StandardScaler()
        X_train_normalized = scaler.fit_transform(X_train)
        X_test_normalized = scaler.transform(X_test)
        
        # Treinar o modelo
        _, w = PLA(X_train_normalized, y_train, max_it)
        
        # Fazer previs√µes
        y_pred = classify(X_test_normalized, w)
        
        # Avaliar o modelo
        accuracy = np.mean(y_pred == y_test)
        accuracies.append(accuracy)
        
    return np.mean(accuracies), np.std(accuracies)
```

```python
# Executar valida√ß√£o cruzada estratificada
mean_accuracy_f0, std_accuracy_f0 = valida_cruzada_estratificada(X_treino_f0, y_treino_f0, max_it=10000)
mean_accuracy_f1, std_accuracy_f1 = valida_cruzada_estratificada(X_treino_f1, y_treino_f1, max_it=1000)
mean_accuracy_f4, std_accuracy_f4 = valida_cruzada_estratificada(X_treino_f4, y_treino_f4, max_it=1000)

print(f"Precis√£o m√©dia do classificador f0: {mean_accuracy_f0:.4f} ¬± {std_accuracy_f0:.4f}")
print(f"Precis√£o m√©dia do classificador f1: {mean_accuracy_f1:.4f} ¬± {std_accuracy_f1:.4f}")
print(f"Precis√£o m√©dia do classificador f4: {mean_accuracy_f4:.4f} ¬± {std_accuracy_f4:.4f}")
```

```python
# Normalizar os dados de treino
scaler_f0 = StandardScaler()
X_treino_f0_normalized = scaler_f0.fit_transform(X_treino_f0)

scaler_f1 = StandardScaler()
X_treino_f1_normalized = scaler_f1.fit_transform(X_treino_f1)

scaler_f4 = StandardScaler()
X_treino_f4_normalized = scaler_f4.fit_transform(X_treino_f4)


# Treinamento dos classificadores
# Classificador f_0: d√≠gito 0 contra 1, 4, 5
_, w_f0 = PLA(X_treino_f0_normalized, y_treino_f0, 10000)

# Classificador f_1: d√≠gito 1 contra 0, 4, 5
_, w_f1 = PLA(X_treino_f1_normalized, y_treino_f1, 1000)

# Classificador f_4: d√≠gito 4 contra 0, 1, 5
_, w_f4 = PLA(X_treino_f4_normalized, y_treino_f4, 1000)

modelos_perceptron = [w_f0, w_f1, w_f4]
```

```python
# Normalizar dados de teste
X_teste_normalized_f0 = scaler_f0.transform(X_teste)
X_teste_normalized_f1 = scaler_f1.transform(X_teste)
X_teste_normalized_f4 = scaler_f4.transform(X_teste)

# Fazer previs√µes usando cada classificador
y_pred_f0 = classify(X_teste_normalized_f0, w_f0)
y_pred_f1 = classify(X_teste_normalized_f1, w_f1)
y_pred_f4 = classify(X_teste_normalized_f4, w_f4)

# Crie a lista de scalers
scaler_modelos = [scaler_f0, scaler_f1, scaler_f4]

# Iterar sobre cada exemplo de teste
y_pred_pla = []
for i in range(X_teste.shape[0]):
    x = X_teste[i]
    y_pred = classificar_digito(x, modelos_perceptron, scaler_modelos, model_type='perceptron')
    y_pred_pla.append(y_pred)

y_pred_pla = np.array(y_pred_pla)

p_conf_matrix = confusion_matrix(y_teste, y_pred_pla)
p_class_report = classification_report(y_teste, y_pred_pla, target_names=['0', '1', '4', '5'], zero_division=1)

print(f"\nMatriz de Confus√£o:\n{p_conf_matrix}")
print(f"\nRelat√≥rio de Classifica√ß√£o:\n{p_class_report}")
```

```python
# foi o unico grafico que ta dando algo parecido com a matriz

def plot_decision_boundary(X, y, w, title):
    plt.figure(figsize=(8, 6))

    # Plotar os dados
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired, edgecolor='k', s=50)
    
    # Calcular a fronteira de decis√£o
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01), np.arange(y_min, y_max, 0.01))
    Z = classify(np.c_[xx.ravel(), yy.ravel()], w).reshape(xx.shape)
    
    # Plotar a fronteira de decis√£o
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.Paired)
    plt.title(title)
    plt.xlabel('Intensidade')
    plt.ylabel('Simetria')
    plt.show()

# Plotar fronteiras de decis√£o para cada classificador
plot_decision_boundary(X_treino_f0_normalized, y_treino_f0, w_f0, 'Fronteira de Decis√£o para f_0')
plot_decision_boundary(X_treino_f1_normalized, y_treino_f1, w_f1, 'Fronteira de Decis√£o para f_1')
plot_decision_boundary(X_treino_f4_normalized, y_treino_f4, w_f4, 'Fronteira de Decis√£o para f_4')
```

```python
# foi o unico grafico que ta dando algo parecido com a matriz e nem ta t√£o parecido

def plot_classification_results(models, X_teste, y_teste, title):

    plt.figure(figsize=(10, 6))
    
    # Definindo os limites do gr√°fico
    x_min, x_max = X_teste[:, 0].min() - 1, X_teste[:, 0].max() + 1
    y_min, y_max = X_teste[:, 1].min() - 1, X_teste[:, 1].max() + 1
    
    # Criando uma malha de pontos para plotar a fronteira de decis√£o
    xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                         np.arange(y_min, y_max, 0.01))
    grid_points = np.c_[xx.ravel(), yy.ravel()]
    
    # Classificando todos os pontos da malha
    Z = np.array([classificar_digito(x, models, model_type='logistic') for x in grid_points])
    Z = Z.reshape(xx.shape)
    
    # Plotando a fronteira de decis√£o
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.RdYlBu)
    
    # Plotando os dados de teste com a cor dos r√≥tulos verdadeiros
    scatter = plt.scatter(X_teste[:, 0], X_teste[:, 1], c=y_teste, cmap=plt.cm.RdYlBu, edgecolors='k')
    
    # Adicionando t√≠tulo e r√≥tulos dos eixos
    plt.title(title)
    plt.xlabel('Intensidade')
    plt.ylabel('Simetria')
    plt.colorbar(scatter)
    
    plt.show()

# Exemplo de uso da fun√ß√£o com os modelos e dados normalizados
plot_classification_results([modelo_f0, modelo_f1, modelo_f4], X_teste_normalized_f0, y_teste, title="Fronteira de Decis√£o e Dados Normalizados")
```
