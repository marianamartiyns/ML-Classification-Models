# Classificação dos dígitos 1x5;

### Importando Bibliotecas

```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
```

### Dados Reduzidos (Treino e Teste)

```python
#primeiros resultados arq de treino
df_train_redu = pd.read_csv('dataset_digits/train_redu.csv', delimiter=';')
print("Primeiras linhas do arquivo train_redu.csv:")
print(df_train_redu.head())

#primeiros resultados arq de teste
df_test_redu = pd.read_csv('dataset_digits/test_redu.csv', delimiter=';')
print("\nPrimeiras linhas do arquivo test_redu.csv:")
print(df_test_redu.head())

#Quanto menor o valor de simetria melhor
```

### 1. train1x5 e test1x5

Realizar um filtro nos dados dos arquivos train_redu.csv e test_redu.csv, deixando apenas as imagens com valores 1 ou 5 na coluna label, construindo as instâncias train1x5 e test1x5;

```python
train1x5 = df_train_redu[(df_train_redu['label'] == 1) | (df_train_redu['label'] == 5)]
test1x5 = df_test_redu[(df_test_redu['label'] == 1) | (df_test_redu['label'] == 5)]

print("Primeiros resultados de train1x5:")
train1x5
```

```python
# Verificar a distribuição das labels
print("\nDistribuição de rótulos nos dados de Train:")
print(train1x5['label'].value_counts())

print("\nDistribuição de rótulos nos dados de Test:")
print(test1x5['label'].value_counts())
```

### 2. Plotar os dados de train1x5 e test1x5

Plotar os dados de train1x5 e test1x5 em um gráfico de duas dimensões (intensidade X simetria). Dados com 𝑙𝑎𝑏𝑒𝑙 = 1 plotar de azul e dados com 𝑙𝑎𝑏𝑒𝑙 = 5 plotar de vermelho;

```python
# Separar as classes no conjunto de treino
train1 = train1x5[train1x5['label'] == 1]
train5 = train1x5[train1x5['label'] == 5]

# Separar as classes no conjunto de teste
test1 = test1x5[test1x5['label'] == 1]
test5 = test1x5[test1x5['label'] == 5]

# Plot dos dados de treino
plt.figure(figsize=(16, 6))

plt.subplot(1, 2, 1)  # Gráfico à esquerda
plt.scatter(train1['intensidade'], train1['simetria'], color='blue', label='Dígito 1 (Treino)')
plt.scatter(train5['intensidade'], train5['simetria'], color='red', label='Dígito 5 (Treino)')
plt.xlabel('Intensidade')
plt.ylabel('Simetria')
plt.title('Classificação binária entre os dígitos 1 e 5 (Treino)')
plt.legend()
plt.grid(True)

# Plot dos dados de teste
plt.subplot(1, 2, 2)  # Gráfico à direita
plt.scatter(test1['intensidade'], test1['simetria'], color='blue', label='Dígito 1 (Teste)')
plt.scatter(test5['intensidade'], test5['simetria'], color='red', label='Dígito 5 (Teste)')
plt.xlabel('Intensidade')
plt.ylabel('Simetria')
plt.title('Classificação binária entre os dígitos 1 e 5 (Teste)')
plt.legend()
plt.grid(True)

# Mostrar ambos os gráficos
plt.tight_layout()
plt.show()
```

### 3. Treinar os três classificadores com os dados de train1x5 e Testa-los com test1x5.

- Construir o vetor 𝑋 = {(𝑖𝑛𝑡𝑒𝑛𝑠𝑖𝑑𝑎𝑑𝑒, 𝑠𝑖𝑚𝑒𝑡𝑟𝑖𝑎)} para toda imagem de train1x5. Atribuir o valor alvo 𝑦 = +1 para a imagem com valor 𝑙𝑎𝑏𝑒𝑙 = 1 e 𝑦 = −1 para 𝑙𝑎𝑏𝑒𝑙 = 5;

- Gerar a matriz de confusão e os relatórios de eficácia de classificação de cada 
classificadro.

```python
# importando classes da pasta 'modelos'

from modelos.Perceptron import PLA, classify
from modelos.RegressaoLinear import RegressaoLinear
from modelos.RegressaoLogistica import RegressaoLogistica_
```

```python
# Separar os dados de treino e teste
X_train = train1x5[['intensidade', 'simetria']].values
y_train = train1x5['label'].apply(lambda x: 1 if x == 1 else -1).values
print(f"Número de elementos no conjunto de treino: {len(X_train)}")

X_test = test1x5[['intensidade', 'simetria']].values
y_test = test1x5['label'].apply(lambda x: 1 if x == 1 else -1).values
print(f"Número de elementos no conjunto de teste: {len(X_test)}")
```

#### Função para plote dos três classificadores

```python
# Função para plotar os resultados
def plot_results(X, y_pred, weights, title, xlabel, ylabel, color_ticks, color_label, xlim=None, ylim=None):
    plt.figure(figsize=(8, 6))
    
    # Plotar dados
    scatter = plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap=plt.cm.RdYlBu, edgecolor='k', s=40, alpha=0.8)
    
    # Adicionar a reta de decisão
    x_values = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
    y_values = - (weights[1] * x_values + weights[0]) / weights[2]
    plt.plot(x_values, y_values, 'k--', label='f(x)')
    
    # Definir limites para os eixos se fornecidos
    if xlim is not None:
        plt.xlim(xlim)
    if ylim is not None:
        plt.ylim(ylim)
    
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.colorbar(scatter, ticks=color_ticks, label=color_label)
    plt.legend()
    plt.show()
```

#### 3.1 Perceptron

```python
# Treinar o Perceptron
it, w = PLA(X_train, y_train, 100000)
y_pred = classify(X_test, w)

y_train_pred = classify(X_train, w)
train_conf_matrix = confusion_matrix(y_train, y_train_pred)
train_class_report = classification_report(y_train, y_train_pred, target_names=['Classe -1', 'Classe 1'])
```

```python
# Avaliar o desempenho no conjunto de teste
y_test_pred = classify(X_test, w)
test_conf_matrix = confusion_matrix(y_test, y_test_pred)
test_class_report = classification_report(y_test, y_test_pred, target_names=['Classe -1', 'Classe 1'])

# Exibir informações do teste
print(f"\nMatriz de Confusão (Teste):\n {test_conf_matrix}")
print(f"\nRelatório de Classificação (Teste):\n{test_class_report}")

# Plotar resultados do treino
plot_results(X_test, y_test_pred, w, title="Resultados do Teste - Perceptron", 
             xlabel="Intensidade", ylabel="Simetria", 
             color_ticks=[-1, 1], color_label="Classe",
             xlim=(40, 120), ylim=(40, 160))
```

#### 3.2 Regressão Linear

```python
lr = RegressaoLinear()
lr.fit(X_train, y_train)

# Fazer previsões nos dados de treino
lr_train_predictions = lr.predict(X_train)
lr_y_train_pred = np.where(lr_train_predictions > 0, 1, -1) # Convertendo previsões contínuas em categorias
```

```python
# Fazer previsões nos dados de teste
lr_test_predictions = lr.predict(X_test)
lr_y_test_pred = np.where(lr_test_predictions > 0, 1, -1) # Convertendo previsões contínuas em categorias

# Avaliar o desempenho no conjunto de teste
lr_test_conf_matrix = confusion_matrix(y_test, lr_y_test_pred)
lr_test_class_report = classification_report(y_test, lr_y_test_pred, target_names=['Classe -1', 'Classe 1'])

# Exibir resultados do teste
print("Resultados do Teste:")
print(f"\nMatriz de Confusão (Teste):\n {lr_test_conf_matrix}")
print(f"\nRelatório de Classificação (Teste):\n{lr_test_class_report}")

# Plotar resultados do teste
plot_results(X_test, lr_y_test_pred, lr.getW(), title="Resultados do Teste - Regressão Linear", 
             xlabel="Intensidade", ylabel="Simetria", color_ticks=[-1, 1], color_label="Classe")
```

#### 3.3 Regressão Logística

```python
# Padronizar as características 
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Adicionar o termo de bias (intercepto) aos dados
X_train_bias = np.hstack([np.ones((X_train.shape[0], 1)), X_train])
X_test_bias = np.hstack([np.ones((X_test.shape[0], 1)), X_test])
```

```python
# Executa o ajuste dos dados com Regressão Logística
classifier = RegressaoLogistica_(eta=0.001, tmax=1500) #Diminui a eta (taxa de aprendizado) pois estava dando overfitting e por isso tava classificando errado
classifier.fit(X_train_bias, y_train)

# Predições para o conjunto de teste
y_pred = classifier.predict(X_test_bias)
```

```python
# Avaliar o desempenho
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred, target_names=['5', '1'])

print(f"\nMatriz de Confusão (Teste):\n {conf_matrix}")
print(f"\nRelatório de Classificação (Teste):\n{class_report}")

# Computa o erro dentro da amostra (Ein)
errorIN = np.mean(y_pred != y_test) * 100
print(f"Error in: {errorIN:.2f}%")

# Plotar resultados do teste
plot_results(X_test, y_test_pred, classifier.getW(), title="Resultados do Teste - Regressão Logística", 
             xlabel="Intensidade", ylabel="Simetria", color_ticks=[-1, 1], color_label="Classe")
```
